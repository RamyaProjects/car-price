Feature Engineering

```python
#checking for null data
df.isnull().sum()
```

```python
#checking for duplicates
df.duplicated().sum()
```

```python
df.head()
```

```python
#converting categorical data to numerical 
df=pd.get_dummies(df,columns=df.select_dtypes(include=['object','category']).columns.tolist(),drop_first=True)
df.head()
```

```python
#explicity mapping False to 0 and True to 1
bool_cols=df.select_dtypes(include='bool').columns.tolist()
for col in bool_cols:
    df[col]=df[col].map({False:0,True:1})
df.head()
```

```python
#defining the features and traget variable
X=df.drop('Price',axis=1)
y=df['Price']
y.shape
```

```python
#splitting the dataset to train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print("No of entries in test data",X_test.shape[0])
print("No of entries in train data",X_train.shape[0])
```

```python
df.head()
```

```python
#normalizing the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
col=['Mileage']
X_train[col]=scaler.fit_transform(X_train[col])
X_test[col]=scaler.transform(X_test[col])
```

```python
X_train.head()
```

LinearRegression using sklearn

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#creating an function to evaluate metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
def metrics_evaluation(model,predictors,target):
    #predicting values of  data
    y_pred=model.predict(predictors)
    #evaluating Metric
   
    r2 = r2_score(target, y_pred)  # to compute R-squared
    rmse = np.sqrt(mean_squared_error(target, y_pred))  # to compute RMSE
    mae = mean_absolute_error(target, y_pred)  # to compute MAE
    mse=mean_squared_error(target,y_pred)
    
    perf=pd.DataFrame({
             "RMSE": rmse,
            "MAE": mae,
            "R-squared": r2,
            "MSE":mse},index=[0])
    return perf
```

```python
metrics_evaluation(model,X_test,y_test)
```

```python
#plotting the residuals and predcted values
import matplotlib.pyplot as plt

# Residuals plot
y_pred=model.predict(X_test)
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted')
plt.show()
```

Linear Regression using statsmodel


```python
import statsmodels.api as sm
X_train=sm.add_constant(X_train)
model1=sm.OLS(y_train,X_train).fit()
print(model1.summary())
```

```python
#checking for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def check_vif(predictors):
    vif=pd.DataFrame()
    vif['feature']=predictors.columns
    vif['VIF']=[variance_inflation_factor(predictors.values,i)
            for i in range(predictors.shape[1])]
    print(vif)

```

```python
#checking multicollinearity
check_vif(X_train)
```

```python
observations:The data contains high multicollinearity
```

Recursive Function Elimination

```python
#from sklearn import datasets
from sklearn.feature_selection import RFE

# Initialize the model- Linear Regression
model = LinearRegression()
# Initialize RFE with the model and the desired number of features to keep
rfe = RFE(estimator=model, n_features_to_select=20)
# Fit RFE to the data
rfe = rfe.fit(X_train, y_train)
# Print which features were selected (True = selected, False = removed)
print("Selected features:",rfe.support_)
# Rank the features (1 is the most important, higher numbers indicate 
print("Feature rankings:",rfe.ranking_)
selected_features = X_train.columns[rfe.support_]
print("Selected features (by name):", selected_features)
```

```python
X_train1=X_train[selected_features]
X_test1=X_test[selected_features]
```

```python
#checking multi collinearity
check_vif(X_train1)
```

```python
# Checking the effect of dropping the columns showing high multicollinearity on model performance 
def check_multicollinear_columns(predictors,target,high_vif_cols):
    rmse=[]
    adj_r2=[]
    for cols in high_vif_cols:
        #x_train = predictors[[col for col in predictors.columns if col not in high_vif_cols]]
        x_train = predictors.loc[:, ~predictors.columns.str.startswith(cols)]
        olsmodel = sm.OLS(target, x_train).fit()
        rmse.append(np.sqrt(olsmodel.mse_resid))
        adj_r2.append(olsmodel.rsquared_adj)
    err_val = pd.DataFrame(
            {
                "col": high_vif_cols,
                "Adj. R-squared after_dropping col": adj_r2,
                "RMSE after dropping col": rmse,
            }).sort_values(by="Adj. R-squared after_dropping col", ascending=False)
    err_val.reset_index(drop=True, inplace=True)
    
    return err_val
```

```python
cols=['Year','Engine_Size']
check_multicollinear_columns(X_train1,y_train,cols)
```

```python
X_train2=X_train1.drop('Engine_Size',axis=1)
X_test2=X_test1.drop('Engine_Size',axis=1)
```

```python
#checking vif for updated dataset
check_vif(X_train2)
```

```python
X_train2=sm.add_constant(X_train2)
olsmodel=sm.OLS(y_train,X_train2).fit()
print(olsmodel.summary())
```

```python
#removing high p values
cols=['Brand_Honda','Brand_Kia','Brand_Mercedes','Model_5 Series','Model_Camry','Model_Civic','Model_Explorer','Model_GLA','Model_Optima','Model_RAV4','Model_Rio','Model_Tucson','Model_X5']
X_train3=X_train2.drop(cols,axis=1)
X_test3=X_test2.drop(cols,axis=1)
```

```python
X_train3=sm.add_constant(X_train3)
X_test3=sm.add_constant(X_test3)
model4=sm.OLS(y_train,X_train3).fit()
print(model4.summary())
```
